```python
import numpy as np
from sklearn.metrics import classification_report,precision_score,recall_score,f1_score,precision_recall_fscore_support

test_y = np.load('./ChemProt/ground_truth.npy')
true_label = np.argmax(test_y, axis=1)
precision_list_weighted_average =[]
recall_list_weighted_average =[]
f1_list_weighted_average =[]
precision_list_macro_average =[]
recall_list_macro_average =[]
f1_list_macro_average =[]
for i in range(1,6):
    allprob = np.load('./ChemProt/version_'+str(i)+'/testset_BiGRU-HAN-WLGCN_HEAD.npy')
    predict_label = np.argmax(allprob, axis=1)
    result_tmp = precision_recall_fscore_support(true_label,predict_label,average='weighted')
    precision_list_weighted_average.append(result_tmp[0])
    recall_list_weighted_average.append(result_tmp[1])
    f1_list_weighted_average.append(result_tmp[2])

    allprob = np.load('./ChemProt/version_'+str(i)+'/testset_BiGRU-HAN_HEAD.npy')
    predict_label = np.argmax(allprob, axis=1)
    result_tmp = precision_recall_fscore_support(true_label,predict_label,average='weighted')
    precision_list_weighted_average.append(result_tmp[0])
    recall_list_weighted_average.append(result_tmp[1])
    f1_list_weighted_average.append(result_tmp[2])

    allprob = np.load('./ChemProt/version_'+str(i)+'/testset_WLGCN_HEAD.npy')
    predict_label = np.argmax(allprob, axis=1)
    result_tmp = precision_recall_fscore_support(true_label,predict_label,average='weighted')
    precision_list_weighted_average.append(result_tmp[0])
    recall_list_weighted_average.append(result_tmp[1])
    f1_list_weighted_average.append(result_tmp[2])


    allprob = np.load('./ChemProt/version_'+str(i)+'/testset_BiGRU-HAN-WGCN_HEAD.npy')
    predict_label = np.argmax(allprob, axis=1)
    result_tmp = precision_recall_fscore_support(true_label,predict_label,average='weighted')
    precision_list_weighted_average.append(result_tmp[0])
    recall_list_weighted_average.append(result_tmp[1])
    f1_list_weighted_average.append(result_tmp[2])

    allprob = np.load('./ChemProt/version_'+str(i)+'/testset_WGCN_HEAD.npy')
    predict_label = np.argmax(allprob, axis=1)
    result_tmp = precision_recall_fscore_support(true_label,predict_label,average='weighted')
    precision_list_weighted_average.append(result_tmp[0])
    recall_list_weighted_average.append(result_tmp[1])
    f1_list_weighted_average.append(result_tmp[2])
```
